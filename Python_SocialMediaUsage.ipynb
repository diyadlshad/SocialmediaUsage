# initalizing every possible package for data set
#### such as numpy, matplotlib for polting based on values
#### pandas for reading dataset and work with.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt


url='https://raw.githubusercontent.com/Ballen-jaza/SocialmediaUsage/main/socialmediausage.csv'

mydata = pd.read_csv(url)

mydata
```

```python
import seaborn as sns

sns.boxplot(x="gender", y="age", data=mydata)
```

```python
sns.pairplot(mydata)
```

# mean of the dataset

```python
x = st.skewnorm.rvs(-4, size=10000000)
x[0:10]
```

```python
fig, ax = plt.subplots()
plt.axvline(x = x.mean(), color='orange')
_ = plt.hist(x, color = 'darkblue')
```

# Median

```python
np.median(x) 
```

```python
fig, ax = plt.subplots()
plt.axvline(x = np.mean(x), color='darkblue')
plt.axvline(x = np.median(x), color='gold')
_ = plt.hist(x, color = 'lightgray')
```
